```python
df.drop(['id'],axis=1,inplace=True)
```

```python
plt.figure(figsize=(15,5))
sns.countplot(data=df,x='diagnosis')
```

```python
df['diagnosis']=df['diagnosis'].replace({'M':0,'B':1})
plt.figure(figsize=(15,5))
sns.countplot(data=df,x='diagnosis')
```

```python
df['diagnosis'].value_counts().to_frame()
```

```python
from sklearn.utils import resample
df_majority=df[df['diagnosis']==1]
df_minority=df[df['diagnosis']==0]
df_majority_downsample=resample(df_majority,n_samples=212,replace=False,random_state=42)
df_df=pd.concat([df_majority_downsample,df_minority])
df_df['diagnosis'].value_counts().to_frame()
```

```python
x=df.drop(['diagnosis'],axis=1)
y=df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
# rf = RandomForest()
# dtf = DecisionTreeClassifier()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.score(x_train, y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,recall_score,precision_score,accuracy_score,f1_score,ConfusionMatrixDisplay
precision_score(y_test,y_pred)
```

```python
from sklearn import svm
svc = svm.SVC()
svc.fit(x_train, y_train)
```

```python
y_pred = svc.predict(x_test)
```

```python
from sklearn.metrics import precision_score
precision_score(y_test, y_pred)
```
